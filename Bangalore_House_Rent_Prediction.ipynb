### **Final PreProcessing**

```python
df.head()
```

```python
new_df = pd.get_dummies(df,columns=['location'],drop_first=True)
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = new_df.drop(columns=['price'])
y = new_df['price']
```

```python
X.sample(5)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

### **Algo Time**

Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
yp = lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, yp)
print(f"R2 Score => {r2}")
```

ANN

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(70,activation='relu',input_dim=X_train.shape[1]))
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
r2_score(y_test,y_pred)
```

XGBOOST

```python
import xgboost as xgb
```

```python
xgb = xgb.XGBRegressor()
xgb.fit(X_train_scaled, y_train)
```

```python
y_pr = xgb.predict(X_test_scaled)
```

```python
r2 = r2_score(y_test, y_pr)
```

```python
print("R2 Score:", r2)
```

Gradient Boost

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(X_train_scaled, y_train)
y_pre = gb.predict(X_test_scaled)
r2 = r2_score(y_test, y_pre)
print("R2 Score:", r2)
```
